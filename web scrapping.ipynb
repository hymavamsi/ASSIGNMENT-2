# IMPORTING THE LIBRARIES

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# GET THE URL

```python
url= "http://quotes.toscrape.com/page/1/"
```

```python
response= requests.get(url)
```

```python
response= response.content
```

# PARSE THE HTML CONTENT

```python
soup= BeautifulSoup(response,"html.parser")
```

```python
soup
```

# EXTRACT DATA FROM PAGE 1

```python
division= soup.find("div")
division
```

```python
division1= division.find_all("div",class_= "quote")
division1
```

```python
for i in division1:
    quotes=i.find("span",class_="text").text
    print(quotes)
```

```python
for i in division1:
    author=i.find("small",class_="author").text
    print(author)
```

```python
for i in division1:
    tags1=i.find("meta",class_="keywords")
    tags2=tags1.attrs["content"]
    print(tags2)
```

# SCRAP DATA FROM ALL THE PAGES

```python
data=[]

for pages in range(1,11):
    url= f"http://quotes.toscrape.com/page/{pages}/"
    response= requests.get(url)
    response= response.content
    soup= BeautifulSoup(response,"html.parser")
    division= soup.find("div")
    division1= division.find_all("div",class_= "quote")
    
    for i in division1:
        quotes=i.find("span",class_="text").text
        author=i.find("small",class_="author").text
        tags1=i.find("meta",class_="keywords")
        tags2=tags1.attrs["content"]
        data.append([quotes,author,tags2])
        
    
    
```

```python
data
```

# IMPORTING THE SCRAPED DATA INTO DATAFRAME

```python
df= pd.DataFrame(data,columns=["QUOTES","AUTHOR","TAGS"])
df
```

# LOADING THE DATA INTO  A CSV FILE

```python
df.to_csv("web_scrapping_project.csv")
```
